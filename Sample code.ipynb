```python
from ultralytics import YOLO
```

```python
!pip uninstall -y albumentations
```

```python
size= 1440
yaml= '/content/drive/MyDrive/Data/Parasite_Detection/data.yaml'
```

```python
model = YOLO('yolo11n.pt')
model.train(data=yaml, batch=16, val=False, epochs=57, close_mosaic=0, imgsz=size, amp=True, project='/content/drive/MyDrive/models', name='Parasite_Detection', exist_ok=True, rect=False, erasing=0.0, flipud=0.5, hsv_h=0.0, hsv_s=0.5, hsv_v=0.1, mosaic=1.0, scale=0.0)
```

```python
metrics=model.val(data=yaml, split='val', conf=0.15, rect=True, cache=False, project='/content/runs/Parasite_Detection')
```
